```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path='/content/train 2.1.csv'
df= pd.read_csv(path)
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['Postal Code'])
```

```python
df.duplicated().sum()
df= df.drop_duplicates()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')

```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
df['profit margin'] = df['Profit'] / df['Sales']
```

## Total Revenue & **Profit**


```python
df['total sales'] = df['Sales'].sum()
df['total profit'] = df['Profit'].sum()
```

```python
df.groupby('Category')[['Sales','Profit']].sum()
```

```python
df.groupby('Region')['Profit'].sum().sort_values(ascending=True)
```

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
```

```python
monthly_sales = df.groupby(['Year','Month'])['Sales'].sum()
monthly_sales.plot()
plt.show()
```

```python
df.groupby('Segment')['Sales'].sum()
```
